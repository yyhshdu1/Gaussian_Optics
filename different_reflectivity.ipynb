```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Define transmission function with different reflectivity

```python
def It(phi, r1, r2):
    R1 = r1**2
    R2 = r2**2
    It = 1 / (1 + ((r1-r2)**2+4*r1*r2*(np.sin(phi/2))**2)/((1-R1)*(1-R2)))
    return It

def R_to_r(R):
    return np.abs(np.sqrt(R))
```

```python
phi = np.arange(3, 4*np.pi+1, 0.001)
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(phi, It(phi,0.995,0.995), '-', ms = 5)
ax.set_xlabel('phase', fontsize=15)
ax.set_ylabel('I / I0',fontsize=15)
```

```python
R2 = np.arange(0.99, 1, 0.0001)
r2 = R_to_r(R2)
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(r2, It(2*np.pi,R_to_r(0.9995),r2), '-', ms = 5, label = "r1 = 99.975% (R = 99.95%)")
# ax.hlines(y=0.3,xmin=R_to_r(0.99),xmax=R_to_r(0.9995), colors= 'black',linestyle='dashed')
ax.set_xlabel('r2', fontsize=15)
ax.set_ylabel('I / I0',fontsize=15)
ax.set_ylim([0,1.01])
ax.set_xlim([R_to_r(0.99),1.0])
ax.set_title("Relative Transmission Power when Resonance")
ax.legend()
```

```python
R2 = np.arange(0.999, 1, 0.00001)
r2 = R_to_r(R2)
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(r2, It(2*np.pi,R_to_r(0.99925),r2), '-', ms = 5, label = "r1 = 0.9996 (R = 99.925%)")
ax.vlines(x = R_to_r(0.9995), ymin=0, ymax=1,colors= 'black',linestyle='dashed')
ax.set_xlabel('r2', fontsize=15)
ax.set_ylabel('I / I0',fontsize=15)
ax.set_ylim([0,1.01])
ax.set_xlim([R_to_r(0.999),1.0])
ax.set_title("Relative Transmission Power when Resonance")
ax.legend()
```

#### Define transmission power with loss

```python
def It(phi, r1, r2, l):
    R1 = r1**2
    R2 = r2**2
    It = 1 / (1 + ((r1-r2)**2+(1-l)*(2-(1+l)*r1*r2)*r1*r2+4*r1*r2*l*(np.sin(phi/2))**2)/((1-R1)*(1-R2)))
    return It

def R_to_r(R):
    return np.abs(np.sqrt(R))
```

```python
l = np.arange(0.99, 1, 0.0001)
R1 = 0.9995
R21 = 0.999
R22 = 0.9995
R23 = 0.9999
r1 = R_to_r(R1)
r21 = R_to_r(R21)
r22 = R_to_r(R22)
r23 = R_to_r(R23)
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(l, It(2*np.pi,r1,r21,l), 'o-', ms = 5, label = "r2 = 99.95% (R = 99.9%)")
ax.plot(l, It(2*np.pi,r1,r22,l), 'x-', ms = 5, label = "r2 = 99.975% (R = 99.95%)")
ax.plot(l, It(2*np.pi,r1,r23,l), '.-', ms = 5, label = "r2 = 99.995% (R = 99.99%)")
# ax.hlines(y=0.3,xmin=R_to_r(0.99),xmax=R_to_r(0.9995), colors= 'black',linestyle='dashed')
ax.set_xlabel('l', fontsize=15)
ax.set_ylabel('I / I0',fontsize=15)
ax.set_ylim([0,1.01])
ax.set_xlim([R_to_r(0.99),1.0001])
ax.set_title("Relative Transmission Power when Resonance, r1 = 99.975%")
ax.legend()
```
